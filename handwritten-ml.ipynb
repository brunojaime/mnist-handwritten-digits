### Visualización de las imágenes

```python
from matplotlib import colormaps
print(list(colormaps))
plt.imshow(digits.images[0],cmap="BuGn")
```

```python
plt.figure(figsize=(15,10))
muestreo =50
plt.yticks([])
plt.xticks([])
plt.title(f"Imagen y etiqueta de {muestreo} muestras")
for i in range(muestreo):
    plt.subplot(5,10,i+1)
    plt.xticks([])
    plt.yticks([])

    plt.xlabel(digits.target[i])
    plt.imshow(digits.images[i],cmap="BuGn")
plt.show()
```
